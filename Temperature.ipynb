```python

#cell import library
import sklearn
from sklearn.datasets import load_iris ,load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LinearR
from sklearn.metrics import r2_score,mean_absolute_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data and setup
#Credit data
#thank :
#.      github : kongruksiamza
#.      YouTube: kongruksiam studio

dataset = pd.read_csv("https://raw.githubusercontent.com/kongruksiamza/MachineLearning/refs/heads/master/Linear%20Regression/Weather.csv")
X = dataset["MinTemp"].values.reshape(-1,1)
y = dataset["MaxTemp"].values.reshape(-1,1)

#split data for train and teat
(X_train, X_test,y_train, y_test) = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#create model and train model

#create
Predict_temperature_lr = LinearR() # : LinearRegression

#train
Predict_temperature_lr.fit(X_train,y_train)
y_pred = Predict_temperature_lr.predict(X_test)
```

```python
#visualize data
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred, color = "red")
plt.show()
```

```python
#info and Evaluate model
r2 = r2_score(y_test,y_pred)
print(f"mean absolue error : {mean_absolute_error(y_test[0:2],y_pred[0:2])}")
print(f"R-square           : {r2}")
```
